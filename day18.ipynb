# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as tck

#畫出完整的 cos 圖形
fig,ax = plt.subplots()
x = np.linspace(-2,2)
y_cos = np.cos(x*np.pi)
ax.plot(x,y_cos)
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.show()
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.scatter(X,Y,c=np.arctan2(Y,X),s=100,alpha=0.5)
plt.show()
```
